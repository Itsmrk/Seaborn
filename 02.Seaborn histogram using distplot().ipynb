# Seaborn histogram using distplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.distplot(tips_df['size'])
```

```python
sns.distplot(tips_df['tip'])
```

```python
sns.distplot(tips_df['total_bill'])
```

```python
sns.distplot(tips_df['total_bill'],bins= 10)
```

```python
sns.distplot(tips_df['total_bill'],bins= 100)
```

```python
sns.distplot(tips_df['total_bill'],bins= 100,hist = False)
```

```python
sns.distplot(tips_df['total_bill'],bins= 100,kde = False) # kernal density estimater
```

```python
sns.distplot(tips_df['total_bill'],bins= 100,kde = False , rug = True) 
```

```python
sns.distplot(tips_df['total_bill'],fit = norm) 
```

```python
sns.distplot(tips_df['total_bill'],kde = False ,fit = norm) 
```

```python
sns.distplot(tips_df['total_bill'],bins= 100,kde = False , color = 'r') 
```

```python
sns.distplot(tips_df['total_bill'],bins= 100,kde = False , vertical = True) 
```

```python
sns.distplot(tips_df['total_bill'],bins= 100,kde = False , color = 'r',norm_hist = True) 
```

```python
sns.distplot(tips_df['total_bill'],axlabel = "Total Bill") 
```

```python
sns.distplot(tips_df['total_bill'],label = "Total Bill") 
plt.legend()
```

```python
sns.distplot(tips_df['total_bill'],label = "Total Bill") 
plt.title("Histogram of Total Bill")
plt.legend()
```

```python
plt.figure(figsize = (16,9))
sns.set()
sns.distplot(tips_df['total_bill'],label = "Total Bill") 
plt.title("Histogram of Total Bill")
plt.legend()
```

```python
tips_df.total_bill.sort_values()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize = (16,9))
sns.set()
sns.distplot(tips_df['total_bill'],bins = bins) 
plt.xticks(bins)
plt.title("Histogram of Total Bill")
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.set()
sns.distplot(tips_df['total_bill'],hist_kws = {'color':'#DC143c','edgecolor':'#aaff00',
                                               'linewidth':5,'linestyle':'--','alpha': 0.9}) 
```

```python
plt.figure(figsize = (16,9))
sns.set()
sns.distplot(tips_df['total_bill'],hist_kws = {'color':'#DC143c','edgecolor':'#aaff00',
                                               'linewidth':5,'linestyle':'--','alpha': 0.9},
            
            kde = {'color':'#8e00ce','linewidth':8,'linestyle':'--','alpha': 0.9}) 
```

```python
plt.figure(figsize = (16,9))
sns.set()
sns.distplot(tips_df['total_bill'],hist_kws = {'color':'#DC143c','edgecolor':'#aaff00',
                                               'linewidth':5,'linestyle':'--','alpha': 0.9},
            
            kde_kws = {'color':'#8e00ce','linewidth':8,'linestyle':'--','alpha': 0.9},
            rug = True,
            rug_kws = {'color':'#0426d0','edgecolor':'#00dbff',
                                               'linewidth':5,'linestyle':'--','alpha': 0.9}) 
```

```python
plt.figure(figsize = (16,9))
sns.set()
sns.distplot(tips_df['total_bill'],hist_kws = {'color':'#DC143c','edgecolor':'#aaff00',
                                               'linewidth':5,'linestyle':'--','alpha': 0.9},
            
            kde = False,
            fit = norm,
            fit_kws = {'color':'#8e00ce','linewidth':8,'linestyle':'--','alpha': 0.7},
            rug = True,
            rug_kws = {'color':'#0426d0','edgecolor':'#aaff00',
                                               'linewidth':5,'linestyle':'--','alpha': 0.9}) 
```

```python
plt.figure(figsize = (16,9))
sns.set()
sns.distplot(tips_df['total_bill'],hist_kws = {'color':'#DC143c','edgecolor':'#aaff00',
                                               'linewidth':5,'linestyle':'--','alpha': 0.9},
            
            kde = False,
            fit = norm,
            fit_kws = {'color':'#8e00ce','linewidth':8,'linestyle':'--','alpha': 0.7},
            rug = True,
            rug_kws = {'color':'#0426d0','edgecolor':'#aaff00',
                                               'linewidth':5,'linestyle':'--','alpha': 0.9},
            label = "TB")
plt.title("Histogram of Resturant Total Bill",fontsize = 20)
plt.xlabel("Total Bill ", fontsize = 15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.distplot(tips_df['total_bill'],bins= 10,label= "Total Bill")
sns.distplot(tips_df['tip'],bins= 10,label= "Tips")
sns.distplot(tips_df['size'],bins= 10,label= "size")
plt.legend()
```

```python
plt.figure(figsize = (16,9))
sns.distplot(tips_df['total_bill'],bins= 10,label= "Total Bill",kde = False)
sns.distplot(tips_df['tip'],bins= 10,label= "Tips",kde = False)
sns.distplot(tips_df['size'],bins= 10,label= "size",kde = False)
plt.legend()
```
