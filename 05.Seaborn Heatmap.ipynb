# Seaborn Heatmap
### Graphical representation of 2d dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arr_2d = np.linspace(1,5,12).reshape(4,3)
arr_2d
```

```python
sns.heatmap(arr_2d)
```

```python
globalWarming_df = pd.read_csv("Who_is_responsible_for_global_warming.csv")
globalWarming_df
```

```python
globalWarming_df  = globalWarming_df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'], axis=1).set_index('Country Name')
globalWarming_df
```

```python
sns.heatmap(globalWarming_df)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,vmin = 0, vmax = 21)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,cmap = "Accent")
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,cmap = "coolwarm")
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,cmap = "coolwarm", annot = True)
```

```python
arr_2d
```

```python
annot_arr = np.array([['a00','a01','a02'],['a10','a11','a12'],['a20','a21','a22'],['a30','a31','a32']])
annot_arr
```

```python
sns.heatmap(arr_2d,annot = annot_arr ,fmt = "s")
plt.show()
```

```python
plt.figure(figsize=(16,9))
 
annot_kws={'fontsize':10, 
           'fontstyle':'italic',  
           'color':"k",
           'alpha':0.9, 
           'rotation':"vertical",
           'verticalalignment':'center',
           'backgroundcolor':'w'}
 
sns.heatmap(globalWarming_df,cmap = "coolwarm", annot = True, annot_kws= annot_kws)
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globalWarming_df, annot = True, cmap = "coolwarm", linewidth = 4)
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globalWarming_df, annot = True, cmap = "coolwarm", linewidth = 4,linecolor = 'k')
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globalWarming_df, annot = True, cbar = False)
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globalWarming_df, cbar = False , xticklabels = False , yticklabels = False)
```

```python
plt.figure(figsize=(16,9))
cbar_kws = { "orientation":"horizontal", 
            "shrink":1,
            'extend':'min', 
            'extendfrac':0.1, 
            "ticks":np.arange(0,22), 
            "drawedges":True }
sns.heatmap(globalWarming_df, cbar_kws = cbar_kws)
```

```python
plt.figure(figsize = (16,9))
country_code = ['USA', 'GBR', 'IND', 'CHN', 'RUS', 'AUS', 'FRA', 'DEU', 'CAN', 'BRA', 'ARG', 'PAK', 'NPL', 'BGD', 'JPN']
sns.heatmap(globalWarming_df,xticklabels= np.arange(0,15), yticklabels = country_code)
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(globalWarming_df)
ax.set(title = "Heat Map of 'WHO is responsible for global Warming'",
      xlabel = "CO2 emission (metric tons per capita) Per Year",
      ylabel = "Country Name")
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(globalWarming_df)
ax.set(title = "Heat Map of 'WHO is responsible for global Warming'",
      xlabel = "CO2 emission (metric tons per capita) Per Year",
      ylabel = "Country Name")
sns.set(font_scale = 2)
```
