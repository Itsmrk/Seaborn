# Correlation Headmap

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
globalWarming_df = pd.read_csv("Who_is_responsible_for_global_warming.csv")
globalWarming_df
```

```python
globalWarming_df  = globalWarming_df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'], axis=1).set_index('Country Name')
globalWarming_df
```

```python
globalWarming_df.corr()
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(globalWarming_df.corr(), annot = True)
ax.tick_params(size = 10 , color = 'k', labelsize = 10 , labelcolor = 'k')
plt.title("Who is responsible for global warming",fontsize = 15)
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
cancer_dataset
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],
                        columns = np.append(cancer_dataset['feature_names'],['target']))
cancer_df
```

```python
plt.figure(figsize = (25,25))
ax = sns.heatmap(cancer_df.corr(), annot =  True, linewidth = 3)
ax.tick_params(size = 10 , color = 'k', labelsize = 10 , labelcolor = 'k')
plt.title("Heat map of breast cancer",fontsize = 15)
plt.show()
```
